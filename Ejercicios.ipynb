Ejercicio 1

```python
import numpy as np
import matplotlib.pyplot as plt

def regla_trapecio_grafica(f, a, b, n):
    # 1. Crear los nodos x y evaluar la función y
    x = np.linspace(a, b, n + 1)
    y = f(x)
    
    # 2. Calcular el ancho h
    h = (b - a) / n
    
    # 3. Fórmula del Trapecio usando Numpy (muy eficiente)
    # Suma: (h/2) * [y_0 + 2*y_1 + ... + 2*y_{n-1} + y_n]
    integral = (h / 2) * (y[0] + 2 * np.sum(y[1:-1]) + y[-1])
    
    # --- ZONA DE GRÁFICA (Matplotlib) ---
    plt.figure(figsize=(10, 6))
    
    # A. Graficar la función "suave" (real)
    x_suave = np.linspace(a, b, 200) # Muchos puntos para que se vea curva
    y_suave = f(x_suave)
    plt.plot(x_suave, y_suave, 'b-', label='Función Exacta $f(x)$', linewidth=2)
    
    # B. Graficar los Trapecios
    # 'fill_between' rellena el área bajo las líneas rectas que unen los puntos
    plt.fill_between(x, y, alpha=0.3, color='orange', edgecolor='red', label=f'Trapecios (n={n})')
    
    # C. Dibujar las líneas verticales de cada trapecio
    plt.vlines(x, 0, y, colors='red', linestyles='dashed', alpha=0.5)
    
    # Detalles del gráfico
    plt.title(f'Integración Numérica: Regla del Trapecio (n={n})')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Mostrar el resultado en el gráfico
    plt.text(a, max(y_suave)*0.1, f'Área Aprox = {integral:.5f}', fontsize=12, 
             bbox=dict(facecolor='white', alpha=0.8))
    
    plt.show()
    
    return integral
```
