```python
import numpy as np

def f(x):
    return np.exp(-x) * np.cos(5 * x)

def puntomedio(f, a, b):
    integral = ((b-a)/2) * f((a+b)/2)
    return integral

def reglatrapecio(f,a,b):
    integral=((b-a)/2) * (f(a)+f(b))
    return integral

def reglasimpson(f,a,b):
    integral=((b-a)/6) * (f(a)+4*f((a+b)/2)+f(b))
    return integral
```

```python
def punto_medio_comp(f,a,b,n):
    h=(b-a)/n
    x = np.linspace(a,b,n)
    return h*np.sum(f(x))

n1=punto_medio_comp(f,0,2,4)
n2=punto_medio_comp(f,0,2,8)
print("n=4: ",n1)
print("n=8: ",n2)
```

```python
def trapecio_comp(f,a,b,n):
    h=(b-a)/n
    x=np.linspace(a,b,n+1)
    return (h/2)*(f(x[0])+2*np.sum(f(x[1:-1]))+f(x[-1]))

n1=trapecio_comp(f,0,2,4)
n2=trapecio_comp(f,0,2,8)
print("n=4: ",n1)
print("n=8: ",n2)
```

```python
def simpson_comp(f,a,b,n):
    if n%2!=0: return("n debe ser par")
    h=(b - a) / n
    x=np.linspace(a,b,n+1)
    return (h/3)*(f(x[0])+4*np.sum(f(x[1:-1:2]))+2*np.sum(f(x[2:-2:2]))+f(x[-1]))

n1=simpson_comp(f,0,2,4)
n2=simpson_comp(f,0,2,8)
print("n=4: ",n1)
print("n=8: ",n2)
```
