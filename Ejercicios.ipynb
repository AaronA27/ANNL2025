```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import diags
import time

# ==========================================
# 1. SOLUCIÓN EXACTA (Literal a)
# ==========================================
def solucion_exacta(x):
    """
    Calcula la solución exacta y(x) analítica.
    y_h = C1*e^(2x) + C2*e^(3x)
    y_p = 0.1(sin(x) + cos(x))
    """
    # Constantes calculadas analíticamente con las condiciones de frontera y(0)=0, y(1)=0
    # Sistema 2x2 para C1 y C2:
    # C1 + C2 = -0.1
    # C1*e^2 + C2*e^3 = -0.1(sin(1) + cos(1))
    
    A_sys = np.array([[1, 1], [np.exp(2), np.exp(3)]])
    b_sys = np.array([-0.1, -0.1 * (np.sin(1) + np.cos(1))])
    C = np.linalg.solve(A_sys, b_sys)
    C1, C2 = C[0], C[1]
    
    return C1 * np.exp(2*x) + C2 * np.exp(3*x) + 0.1 * (np.sin(x) + np.cos(x))
```
